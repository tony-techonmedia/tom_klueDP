```python
with open('klue_output/transformers/pred/pred-0.json', 'r') as io:
    reader=csv.reader(io,delimiter=' ')
    
    out = []
    for line in reader:
        out.append([int(line[0]),dep_labels[int(line[1])]])
```

```python
out
```

```python
with open('data/03_tok.tsv', 'r') as io:
    reader=csv.reader(io, delimiter='\t')
    idx = 0
    
    with open('data/03_DP.tsv', 'w') as jo:
        writer=csv.writer(jo, delimiter='\t')

        for i, line in enumerate(reader):

            l = len(line)

            if l < 3:
                writer.writerow(line)
            else:
                writer.writerow(line[:-2]+out[idx])
                idx += 1
    
```
